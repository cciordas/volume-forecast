## 1. The Problem: Why Predict Intraday Volume?

Suppose you are an institutional trader who needs to sell 2 million shares of AAPL over the course of a trading day. You cannot dump them all at once -- that would move the market against you. Instead, you split the order across time, executing a little in each 15-minute interval.

But **how much** should you trade in each interval? The answer depends on how volume distributes across the day. If you trade proportionally to the market's natural volume pattern, your trades blend in with the crowd and minimize market impact. This is the idea behind Volume-Weighted Average Price (VWAP) execution.

The catch: you need to **predict** how volume will distribute today, and you need to update that prediction as the day unfolds and new information arrives.

Let's look at what real intraday volume looks like.

````python
# Plot raw intraday volume for a few sample days
bin_labels = [f"{9+i//4}:{(i%4)*15:02d}" for i in range(I)]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: Several individual days overlaid
sample_days = np.random.RandomState(42).choice(range(train_end - 60, train_end), size=8, replace=False)
for d in sorted(sample_days):
    axes[0].plot(range(I), vm[d, :], alpha=0.5, linewidth=1)
axes[0].set_xlabel("15-minute bin")
axes[0].set_ylabel("Volume (shares)")
axes[0].set_title(f"{TICKER}: Intraday Volume on 8 Random Days")
axes[0].set_xticks(range(0, I, 4))
axes[0].set_xticklabels([bin_labels[i] for i in range(0, I, 4)], rotation=45)

# Right: Average intraday pattern (the U-shape)
avg_profile = np.mean(vm[train_end-126:train_end], axis=0)
axes[1].bar(range(I), avg_profile, color='steelblue', alpha=0.7)
axes[1].set_xlabel("15-minute bin")
axes[1].set_ylabel("Average Volume (shares)")
axes[1].set_title(f"{TICKER}: Average Intraday Volume Profile (6-month)")
axes[1].set_xticks(range(0, I, 4))
axes[1].set_xticklabels([bin_labels[i] for i in range(0, I, 4)], rotation=45)

plt.tight_layout()
plt.show()

print(f"First bin (9:30) avg volume: {avg_profile[0]:,.0f}")
print(f"Midday (12:00) avg volume:   {avg_profile[10]:,.0f}")
print(f"Last bin (15:45) avg volume:  {avg_profile[-1]:,.0f}")
print(f"Ratio open/midday:           {avg_profile[0]/avg_profile[10]:.1f}x")
````

Two key observations jump out:

1. **The U-shape is persistent.** Volume is high at the open (9:30), drops during the middle of the day, and surges again near the close (15:45). This pattern appears virtually every day, driven by overnight information processing at the open and portfolio rebalancing at the close.

2. **Day-to-day variation is large.** Even though the shape is stable, the *level* varies enormously -- some days have twice the volume of others, driven by earnings, news, macro events, or options expiration.

A good forecast must capture both the predictable shape and the day-specific level. The simplest approach -- just using the historical average -- captures the shape but misses the level. Let's quantify how much room for improvement exists.

````python
# Show how much daily volume varies day-to-day
daily_totals = np.sum(vm[train_end-126:train_end], axis=1)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(daily_totals / 1e6, 'steelblue', linewidth=1)
axes[0].set_xlabel("Trading Day (relative)")
axes[0].set_ylabel("Daily Volume (millions of shares)")
axes[0].set_title(f"{TICKER}: Daily Volume over 6 Months")
axes[0].axhline(np.mean(daily_totals) / 1e6, color='red', linestyle='--', alpha=0.7, label='Mean')
axes[0].legend()

# Show the prediction error of historical average for one example day
example_day = train_end - 5
hist_avg_example = np.mean(vm[example_day - 21:example_day], axis=0)
actual_example = vm[example_day]

axes[1].bar(np.arange(I) - 0.15, hist_avg_example, width=0.3, alpha=0.7, label='21-day Avg Forecast', color='steelblue')
axes[1].bar(np.arange(I) + 0.15, actual_example, width=0.3, alpha=0.7, label='Actual', color='coral')
axes[1].set_xlabel("15-minute bin")
axes[1].set_ylabel("Volume (shares)")
axes[1].set_title(f"{TICKER}: Historical Average vs Actual (Example Day)")
axes[1].set_xticks(range(0, I, 4))
axes[1].set_xticklabels([bin_labels[i] for i in range(0, I, 4)], rotation=45)
axes[1].legend()

plt.tight_layout()
plt.show()

# Compute baseline MAPE for this day
ape = np.abs(hist_avg_example - actual_example) / np.maximum(actual_example, 100)
print(f"Example day MAPE with historical average: {np.mean(ape)*100:.1f}%")
print(f"Coefficient of variation in daily volume: {np.std(daily_totals)/np.mean(daily_totals)*100:.1f}%")
````

## 2. The Intuition: Three Sources of Information

The key insight of Satish et al. (2014) is that volume in any 15-minute bin can be predicted by combining **three independent sources of information**, each capturing a different time scale:

| Component | Time Scale | What It Captures | Analogy |
|-----------|-----------|-----------------|---------|
| **H** (Historical Average) | Weeks | The stable baseline pattern | "AAPL typically trades 5M shares in the first 15 minutes" |
| **D** (Inter-day ARMA) | Days | Recent trends and mean-reversion in daily volume | "Volume has been running 20% above normal this week" |
| **A** (Intraday ARMA) | Minutes | Real-time same-day conditioning | "Based on what we've seen so far today, volume is running hot" |

The model combines these three components with weights that adapt to the current **volume regime** -- whether today is shaping up to be a high-volume, normal, or low-volume day.

The architecture looks like this:

```
Historical data (21 days) ---> [H] Historical Average ----+
                                                           |
Daily volume series -------> [D] Inter-day ARMA --------+  |
                                                         |  |
Today's observed bins -----> [A] Intraday ARMA -------+  |  |
                                                       |  |  |
                                                       v  v  v
Cumulative volume today --> [Regime Classifier] --> [Dynamic Weights] --> V_hat
```

**Why three components?** Each one fails alone:
- **H alone** misses day-to-day variation (the level shifts we saw above).
- **D alone** captures inter-day dynamics but doesn't adapt to what's happening *today*.
- **A alone** is powerful once the day is underway, but gives no signal at the open (bin 1).

By combining them with dynamic weights, the model starts the day relying on H and D, then gradually shifts weight toward A as intraday observations accumulate and the intraday ARMA becomes well-conditioned.

## 3. Building the Model: Component by Component

### 3.1 Component H: The Historical Average

The simplest forecast is just the average volume in each bin over the recent past. For bin $i$ on day $d$:

$$H_{d,i} = \frac{1}{N_{\text{hist}}} \sum_{k=1}^{N_{\text{hist}}} V_{d-k, i}$$

where $V_{d,i}$ is the actual volume in bin $i$ on day $d$ and $N_{\text{hist}} = 21$ trading days (about 1 month).

This captures the stable U-shaped intraday pattern. Let's compute it and see what it looks like.

````python
# Component H: Historical Average
train_data = vm[:train_end]
hist_avg = compute_historical_average(train_data, params.N_hist)

fig, ax = plt.subplots(figsize=(12, 5))
ax.bar(range(I), hist_avg, color='steelblue', alpha=0.7)
ax.set_xlabel("15-minute bin")
ax.set_ylabel("Volume (shares)")
ax.set_title(f"{TICKER}: Component H -- Historical Average (21-day window)")
ax.set_xticks(range(0, I, 2))
ax.set_xticklabels([bin_labels[i] for i in range(0, I, 2)], rotation=45)
plt.tight_layout()
plt.show()

print(f"H captures the U-shape: open={hist_avg[0]:,.0f}, midday={hist_avg[13]:,.0f}, close={hist_avg[-1]:,.0f}")
print(f"Open/midday ratio: {hist_avg[0]/hist_avg[13]:.2f}x")
````

### 3.2 Seasonal Factors: Removing the Intraday Pattern

Before we can apply time-series models (ARMA) to volume data, we need to remove the strong intraday U-shape pattern. Otherwise, the ARMA model would waste its capacity trying to model what is really just a deterministic daily cycle.

The **seasonal factors** are computed as a long-window (126-day, ~6 months) average per bin:

$$S_i = \frac{1}{N_{\text{seasonal}}} \sum_{k=1}^{N_{\text{seasonal}}} V_{d-k, i}$$

We then **deseasonalize** volume by dividing: $\tilde{V}_{d,i} = V_{d,i} / S_i$

After deseasonalization, the series should fluctuate around 1.0 -- values above 1 mean "more volume than usual for this time of day," and below 1 means less.

Note: the seasonal factors use a *longer* window (126 days) than the historical average H (21 days). The longer window provides a more stable reference for deseasonalization, while the shorter window for H makes it more responsive to recent volume shifts.

````python
# Compute seasonal factors and show deseasonalization
sf = compute_seasonal_factors(train_data, params.N_seasonal)

fig, axes = plt.subplots(1, 3, figsize=(16, 4.5))

# Seasonal factors (the U-shape template)
axes[0].bar(range(I), sf, color='steelblue', alpha=0.7)
axes[0].set_title("Seasonal Factors (126-day avg)")
axes[0].set_xlabel("Bin")
axes[0].set_ylabel("Volume")

# Raw volume for a sample day
sample_day = train_end - 10
axes[1].bar(range(I), vm[sample_day], color='coral', alpha=0.7)
axes[1].set_title(f"Raw Volume (day {sample_day})")
axes[1].set_xlabel("Bin")
axes[1].set_ylabel("Volume")

# Deseasonalized
deseas = vm[sample_day] / sf
axes[2].bar(range(I), deseas, color='seagreen', alpha=0.7)
axes[2].axhline(1.0, color='red', linestyle='--', alpha=0.5, label='Expected mean')
axes[2].set_title("Deseasonalized (raw / seasonal)")
axes[2].set_xlabel("Bin")
axes[2].set_ylabel("Ratio")
axes[2].legend()

plt.tight_layout()
plt.show()

print(f"Deseasonalized series mean: {deseas.mean():.3f} (should be ~1.0)")
print(f"Deseasonalized series std:  {deseas.std():.3f}")

# Verify stationarity with ADF test on concatenated deseasonalized series
concat_deseas = (train_data[-21:] / sf).flatten()
adf_stat, adf_p, *_ = adfuller(concat_deseas, maxlag=10)
print(f"\nADF test on deseasonalized series: stat={adf_stat:.3f}, p={adf_p:.6f}")
print(f"Stationary at 5% level: {adf_p < 0.05}")
````

### 3.3 Component D: Inter-day ARMA

The second component captures how volume in each bin evolves **from day to day**. For each of the 26 bins, we fit an independent ARMA(p, q) model on the daily series of that bin's volume.

**What is ARMA?** An ARMA(p, q) model says that today's value depends on:
- The last $p$ values (the **AR** or autoregressive part): "if volume was high yesterday, it's likely high today"
- The last $q$ forecast errors (the **MA** or moving average part): "if I underpredicted yesterday, I should adjust upward"

$$D_{d,i} = c + \sum_{j=1}^{p} \phi_j \cdot V_{d-j,i} + \sum_{j=1}^{q} \theta_j \cdot \epsilon_{d-j,i}$$

where $c$ is a constant, $\phi_j$ are AR coefficients, $\theta_j$ are MA coefficients, and $\epsilon_{d-j,i}$ are past forecast errors.

We search over all combinations of $p \in \{0,...,5\}$ and $q \in \{0,...,5\}$ and pick the model with the best AICc (corrected Akaike Information Criterion), which balances fit quality against model complexity. The correction (the "c" in AICc) penalizes complex models more heavily when the sample size is small -- important since we're working with only 63 training days.

$$\text{AICc} = \text{AIC} + \frac{2k(k+1)}{n - k - 1}, \quad \text{AIC} = -2\ln L + 2k$$

where $k = p + q + 1$ is the number of parameters and $n$ is the sample size.

````python
# Fit inter-day ARMA models (one per bin)
interday_models = fit_interday_arma(
    train_data, params.N_interday_fit, params.p_max_inter, params.q_max_inter
)

# Report what ARMA orders were selected
orders = []
for i, m in enumerate(interday_models):
    if m is FALLBACK:
        orders.append((i+1, "FALLBACK", "-"))
    else:
        p_sel = m.model.k_ar
        q_sel = m.model.k_ma
        orders.append((i+1, f"ARMA({p_sel},{q_sel})", p_sel + q_sel))

orders_df = pd.DataFrame(orders, columns=["Bin", "Model", "Params"])
print("Selected inter-day ARMA orders per bin:")
print(orders_df.to_string(index=False))
print(f"\nFallback count: {sum(1 for m in interday_models if m is FALLBACK)}")
n_params = [o[2] for o in orders if o[2] != "-"]
print(f"Median AR+MA order: {np.median(n_params):.0f}")
````

````python
# Visualize inter-day ARMA forecasts for the opening bin
bin_idx = 0  # first bin (9:30-9:45)
series = train_data[-63:, bin_idx]

fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(range(len(series)), series, 'steelblue', linewidth=1, label='Actual', alpha=0.8)

# Generate one-step-ahead forecasts
if interday_models[bin_idx] is not FALLBACK:
    forecasts_d = []
    for d in range(20, len(series)):
        try:
            res = interday_models[bin_idx].apply(series[:d])
            fc = res.forecast(steps=1)
            forecasts_d.append((d, float(fc.iloc[0]) if hasattr(fc, 'iloc') else float(fc[0])))
        except Exception:
            pass
    if forecasts_d:
        fc_x, fc_y = zip(*forecasts_d)
        ax.plot(fc_x, fc_y, 'coral', linewidth=1, label='Inter-day ARMA forecast', alpha=0.8)

ax.set_xlabel("Trading Day (within 63-day fitting window)")
ax.set_ylabel("Volume (shares)")
ax.set_title(f"{TICKER} Bin 1 (9:30-9:45): Inter-day ARMA One-Step-Ahead Forecasts")
ax.legend()
plt.tight_layout()
plt.show()
````

### 3.4 Component A: Intraday ARMA

The most powerful component is the intraday ARMA. It operates on **deseasonalized** volume *within* a single day. The key idea: if the first few bins of today are running 30% above normal, the rest of the day is likely to be above normal too.

We fit a single ARMA model on concatenated deseasonalized daily segments from the most recent 21 trading days. At prediction time, we feed in today's observed (deseasonalized) bins and forecast forward:

1. Observe bins 1 through $j$: volumes $V_{d,1}, ..., V_{d,j}$
2. Deseasonalize: $\tilde{V}_{d,i} = V_{d,i} / S_i$ for $i = 1,...,j$
3. Run the ARMA recursion through these observations to build up state
4. Forecast: $\hat{\tilde{V}}_{d,j+1}$ (deseasonalized forecast for next bin)
5. Re-seasonalize: $\hat{V}_{d,j+1} = \hat{\tilde{V}}_{d,j+1} \times S_{j+1}$

This is where the model gets its real-time adaptiveness. Early in the day, the intraday ARMA has little information and its forecast is close to the unconditional mean (~1.0 in deseasonalized space, which re-seasonalizes to the seasonal factor). But by mid-morning, after observing 4-6 bins, the ARMA state is well-conditioned and the forecasts become highly informative.

````python
# Fit intraday ARMA
intraday_model = fit_intraday_arma(
    train_data, sf, params.N_intraday_fit, params.p_max_intra, params.q_max_intra
)

if intraday_model is not FALLBACK:
    print(f"Intraday ARMA order: ({intraday_model.model.k_ar}, {intraday_model.model.k_ma})")
else:
    print("Intraday ARMA: FALLBACK (no valid model found)")

# Demonstrate the intraday conditioning effect on a sample day
demo_day = train_end - 3
actual_day = vm[demo_day]
deseas_day = actual_day / sf

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: show forecasts as we reveal more bins
colors = plt.cm.viridis(np.linspace(0.2, 0.9, 6))
conditioning_bins = [0, 2, 5, 10, 15, 20]

for idx, n_obs in enumerate(conditioning_bins):
    if n_obs == 0:
        observed_deseas = np.array([])
    else:
        observed_deseas = deseas_day[:n_obs]
    
    remaining = I - n_obs
    if remaining > 0 and intraday_model is not FALLBACK:
        fc_deseas = _predict_intraday(intraday_model, observed_deseas, sf, steps=remaining)
        fc_raw = fc_deseas * sf[n_obs:]
        axes[0].plot(range(n_obs, I), fc_raw, color=colors[idx], linewidth=1.5,
                     alpha=0.7, label=f'After {n_obs} bins')

axes[0].plot(range(I), actual_day, 'k-', linewidth=2, alpha=0.5, label='Actual')
axes[0].set_xlabel("Bin")
axes[0].set_ylabel("Volume (shares)")
axes[0].set_title("Intraday ARMA: Forecasts improve with more observations")
axes[0].legend(fontsize=9)

# Right: forecast error shrinks with conditioning
errors_by_obs = []
for n_obs in range(I - 1):
    if n_obs == 0:
        observed_deseas = np.array([])
    else:
        observed_deseas = deseas_day[:n_obs]
    
    target_bin = n_obs  # 0-indexed
    if intraday_model is not FALLBACK:
        fc_deseas = _predict_intraday(intraday_model, observed_deseas, sf, steps=1)
        fc_raw = fc_deseas[0] * sf[n_obs]
    else:
        fc_raw = sf[n_obs]
    error = abs(fc_raw - actual_day[n_obs]) / max(actual_day[n_obs], 100)
    errors_by_obs.append(error)

axes[1].bar(range(len(errors_by_obs)), errors_by_obs, color='steelblue', alpha=0.7)
axes[1].set_xlabel("Target bin (number of prior observations)")
axes[1].set_ylabel("Absolute Percentage Error")
axes[1].set_title("Intraday ARMA: Error decreases as day progresses")

plt.tight_layout()
plt.show()
````

### 3.5 Regime Classification: Adapting to Today's Volume Level

Now we have three forecasts: $H_i$, $D_i$, and $A_i$. How do we combine them?

The naive approach would be fixed weights, but Satish et al. observe that the optimal weights depend on **how much volume has already traded today** relative to historical norms. On a high-volume day, the intraday ARMA signal is more informative (because volume momentum is strong). On a low-volume day, the historical average may be more reliable.

The regime classifier works by:
1. Looking at the cumulative volume through the current bin
2. Comparing it to the historical distribution of cumulative volumes at the same time of day
3. Assigning it to a regime bucket (e.g., "low", "normal-low", "normal-high", "high")

The thresholds are percentiles of the historical cumulative volume distribution. For $K$ regimes, we use equally-spaced percentile cutoffs: $100/K, 200/K, ..., (K-1) \times 100/K$.

For example, with $K=4$ regimes, the cutoffs are at the 25th, 50th, and 75th percentiles.

````python
# Build and visualize regime classifier
n_regimes = 4  # We'll use 4 for illustration (the grid search selects this)
rc = build_regime_classifier(train_data, params.N_regime_window, n_regimes)

# Show the thresholds for a few bins
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: Cumulative volume distribution at mid-day (bin 13)
mid_bin = 13
cumvol_data = np.sum(train_data[-63:, :mid_bin], axis=1)
axes[0].hist(cumvol_data, bins=20, color='steelblue', alpha=0.7, edgecolor='white')
thresholds = rc.thresholds[mid_bin]
colors = ['green', 'orange', 'red']
labels = ['25th pctl', '50th pctl', '75th pctl']
for t, c, l in zip(thresholds, colors, labels):
    axes[0].axvline(t, color=c, linestyle='--', linewidth=2, label=l)
axes[0].set_xlabel("Cumulative Volume through bin 13 (12:45)")
axes[0].set_ylabel("Count (days)")
axes[0].set_title(f"Regime Thresholds at Mid-day ({TICKER})")
axes[0].legend()

# Right: Regime assignment over time for evaluation days
regime_timeline = []
for d in range(train_end, n_days):
    cumvol = 0
    day_regimes = []
    for i in range(I):
        cumvol += vm[d, i]
        r = assign_regime(rc, i + 1, cumvol)
        day_regimes.append(r)
    regime_timeline.append(day_regimes)

regime_arr = np.array(regime_timeline)
im = axes[1].imshow(regime_arr.T, aspect='auto', cmap='RdYlGn_r',
                      interpolation='nearest', vmin=0, vmax=n_regimes-1)
axes[1].set_xlabel("Evaluation Day")
axes[1].set_ylabel("Bin")
axes[1].set_title("Regime Assignment: Each Cell Shows the Volume Regime")
axes[1].set_yticks(range(0, I, 4))
axes[1].set_yticklabels([bin_labels[i] for i in range(0, I, 4)])
plt.colorbar(im, ax=axes[1], label='Regime (0=low, 3=high)')

plt.tight_layout()
plt.show()

# Show how regimes evolve within a single day
print("Example: regime evolution on first evaluation day:")
print(f"  Bins 1-5 regimes:  {regime_arr[0, :5].tolist()}")
print(f"  Bins 10-15 regimes: {regime_arr[0, 10:15].tolist()}")
print(f"  Bins 22-26 regimes: {regime_arr[0, 22:].tolist()}")
````

### 3.6 Dynamic Weights: Putting It All Together

The final forecast for bin $i$ on day $d$ is a weighted combination:

$$\hat{V}_{d,i} = w_H^{(r)} \cdot H_{d,i} + w_D^{(r)} \cdot D_{d,i} + w_A^{(r)} \cdot A_{d,i}$$

where $r$ is the current volume regime, and $w_H^{(r)}, w_D^{(r)}, w_A^{(r)}$ are the regime-specific weights.

**How are the weights optimized?** For each regime, we minimize the Mean Absolute Percentage Error (MAPE) on historical data:

$$\min_{w_H, w_D, w_A \geq 0} \; \frac{1}{N} \sum_{(d,i) \in \text{regime } r} \frac{|w_H H_{d,i} + w_D D_{d,i} + w_A A_{d,i} - V_{d,i}|}{V_{d,i}}$$

Key implementation details:
- **Non-negativity** is enforced via a log-space transformation: we optimize over $\log w$ and exponentiate, so weights are always positive.
- **No sum-to-1 constraint.** Unlike a probability mixture, we allow weights to correct for systematic bias in individual components. The optimal weights might sum to more or less than 1.
- **Multi-restart Nelder-Mead** with 4 starting points (equal, H-dominant, D-dominant, A-dominant) to avoid local minima.

Let's train the full model and examine the learned weights.

````python
# Train the full model (this takes ~90 seconds)
print("Training full model (Model A + Model B)...")
model_a, pct_model = train_full_model(vm, train_end, params)
print("Done!")

# Display the learned weights
print(f"\nNumber of regimes selected: {model_a.regime_classifier.n_regimes}")
print(f"\nPer-regime weights [w_H, w_D, w_A]:")
print(f"{'Regime':<10} {'w_H (Hist)':>12} {'w_D (Inter)':>12} {'w_A (Intra)':>12} {'Sum':>8}")
print("-" * 58)
for r in sorted(model_a.weights.keys()):
    w = model_a.weights[r]
    regime_label = {0: "Low vol", 1: "Med-low", 2: "Med-high", 3: "High vol"}.get(r, str(r))
    print(f"{regime_label:<10} {w[0]:>12.4f} {w[1]:>12.4f} {w[2]:>12.4f} {sum(w):>8.4f}")
````

### 3.7 Model B: Volume Percentage Forecasting

Model A forecasts raw volume (shares). Model B takes a different approach: it forecasts what **fraction** of the day's total volume will trade in the next bin. This is useful for VWAP algorithms that need to allocate a target quantity across the day in proportion to expected volume.

The approach, extending Humphery-Jenner (2011), works in "percentage space":

1. **Historical percentage curve** $p_i^{\text{hist}}$: The average fraction of daily volume in each bin, computed over the training window.

2. **Surprises**: At each bin, the difference between the actual participation rate and the historical one:
$$\text{surprise}_{d,i} = \frac{V_{d,i}}{V_{d,\text{total}}} - p_i^{\text{hist}}$$

3. **Surprise regression**: A linear model that predicts the next surprise from recent surprises:
$$\hat{\delta}_{d,i+1} = \sum_{\ell=1}^{L} \beta_\ell \cdot \text{surprise}_{d,i-\ell+1}$$
where $L$ is selected by time-series cross-validation.

4. **Adjustment with constraints**: The forecast adjusts the scaled historical percentage:
$$\hat{p}_{d,i+1} = \text{scale} \times (p_{i+1}^{\text{hist}} + \text{clamp}(\hat{\delta}, \pm \alpha \cdot p_{i+1}^{\text{hist}}))$$
where $\text{scale}$ rescales remaining percentages to account for observed volume, and $\alpha = 0.10$ limits the maximum deviation.

````python
# Model B: examine the historical percentage curve and surprise regression
fig, axes = plt.subplots(1, 3, figsize=(16, 4.5))

# Left: Historical percentage curve
axes[0].bar(range(I), pct_model.hist_pct * 100, color='steelblue', alpha=0.7)
axes[0].set_xlabel("Bin")
axes[0].set_ylabel("Percentage of Daily Volume (%)")
axes[0].set_title("Historical Volume Percentage Curve")
axes[0].set_xticks(range(0, I, 4))
axes[0].set_xticklabels([bin_labels[i] for i in range(0, I, 4)], rotation=45)

# Middle: Surprise distribution
train_pct = train_data[-63:]
daily_tots = np.sum(train_pct, axis=1)
valid = daily_tots > 100
pct_matrix = train_pct[valid] / daily_tots[valid, np.newaxis]
surprises = pct_matrix - pct_model.hist_pct
axes[1].hist(surprises.flatten(), bins=50, color='coral', alpha=0.7, edgecolor='white')
axes[1].set_xlabel("Surprise (actual_pct - hist_pct)")
axes[1].set_ylabel("Count")
axes[1].set_title("Distribution of Volume Percentage Surprises")
axes[1].axvline(0, color='black', linestyle='--', alpha=0.5)

# Right: Autocorrelation of surprises (motivating the regression)
from numpy import correlate as np_correlate
flat_surp = surprises.flatten()
# Compute lag-1 through lag-5 autocorrelation
acf_vals = []
for lag in range(1, 6):
    s1 = flat_surp[lag:]
    s2 = flat_surp[:-lag]
    acf_vals.append(np.corrcoef(s1, s2)[0, 1])

axes[2].bar(range(1, 6), acf_vals, color='seagreen', alpha=0.7)
axes[2].set_xlabel("Lag")
axes[2].set_ylabel("Autocorrelation")
axes[2].set_title("Surprise Autocorrelation (justifies regression)")
axes[2].axhline(0, color='black', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()

print(f"Model B parameters:")
print(f"  Selected lag L = {pct_model.L}")
print(f"  Beta coefficients: {pct_model.beta}")
print(f"  |Beta| < 1.0: {np.all(np.abs(pct_model.beta) < 1.0)} (bounded, as required)")
print(f"  Positive beta_1: {pct_model.beta[0] > 0} (positive autocorrelation in surprises)")
````

## 4. The Algorithm in Action: Step-by-Step Walkthrough

Let's walk through a complete prediction for one day from the evaluation period. We'll show exactly how the model produces its forecast for each bin, starting from the open.

````python
# Walk through prediction on a single evaluation day
walkthrough_day = train_end + 10  # Pick a day well into the eval period
actual = vm[walkthrough_day]

forecasts = []
components_H = []
components_D = []
components_A = []
regimes_used = []

for target_bin in range(1, I + 1):
    # Build observed volumes (bins before the target)
    observed = {j: vm[walkthrough_day, j-1] for j in range(1, target_bin)}
    
    # Get the full forecast
    v_hat = forecast_raw_volume(model_a, vm, walkthrough_day, target_bin, observed)
    forecasts.append(v_hat)
    
    # Decompose: compute individual components
    tb_idx = target_bin - 1
    H = model_a.hist_avg[tb_idx]
    components_H.append(H)
    
    if model_a.interday_models[tb_idx] is FALLBACK:
        D = H
    else:
        D = _predict_interday_next(model_a.interday_models[tb_idx], vm[:walkthrough_day, tb_idx])
        if np.isnan(D) or np.isinf(D):
            D = H
    components_D.append(D)
    
    current_bin = max(observed.keys()) if observed else 0
    if model_a.intraday_model is FALLBACK:
        A = model_a.seasonal_factors[tb_idx]
    else:
        if current_bin > 0:
            obs_deseas = np.array([observed[j] / model_a.seasonal_factors[j-1] for j in range(1, current_bin+1)])
        else:
            obs_deseas = np.array([])
        steps_ahead = target_bin - current_bin
        fc = _predict_intraday(model_a.intraday_model, obs_deseas, model_a.seasonal_factors, steps=max(steps_ahead, 1))
        A = fc[-1] * model_a.seasonal_factors[tb_idx]
        if np.isnan(A) or np.isinf(A):
            A = model_a.seasonal_factors[tb_idx]
    components_A.append(A)
    
    cumvol = sum(observed.values())
    r = assign_regime(model_a.regime_classifier, current_bin, cumvol)
    regimes_used.append(r)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Top left: Forecast vs Actual
axes[0, 0].plot(range(I), actual, 'ko-', linewidth=2, markersize=4, label='Actual', alpha=0.7)
axes[0, 0].plot(range(I), forecasts, 'rs-', linewidth=2, markersize=4, label='Model A Forecast', alpha=0.7)
axes[0, 0].set_xlabel("Bin")
axes[0, 0].set_ylabel("Volume")
axes[0, 0].set_title("Model A: Forecast vs Actual (Single Day Walkthrough)")
axes[0, 0].legend()
axes[0, 0].set_xticks(range(0, I, 4))
axes[0, 0].set_xticklabels([bin_labels[i] for i in range(0, I, 4)], rotation=45)

# Top right: Three components
axes[0, 1].plot(range(I), components_H, 'b--', linewidth=1.5, label='H (Historical)', alpha=0.7)
axes[0, 1].plot(range(I), components_D, 'r--', linewidth=1.5, label='D (Inter-day)', alpha=0.7)
axes[0, 1].plot(range(I), components_A, 'g-', linewidth=1.5, label='A (Intraday)', alpha=0.7)
axes[0, 1].plot(range(I), actual, 'ko', markersize=3, alpha=0.3, label='Actual')
axes[0, 1].set_xlabel("Bin")
axes[0, 1].set_ylabel("Volume")
axes[0, 1].set_title("Individual Component Forecasts")
axes[0, 1].legend(fontsize=9)

# Bottom left: Per-bin error
errors = np.abs(np.array(forecasts) - actual) / np.maximum(actual, 100) * 100
axes[1, 0].bar(range(I), errors, color='coral', alpha=0.7)
axes[1, 0].set_xlabel("Bin")
axes[1, 0].set_ylabel("Absolute Percentage Error (%)")
axes[1, 0].set_title(f"Per-Bin Error (MAPE = {np.mean(errors):.1f}%)")
axes[1, 0].set_xticks(range(0, I, 4))
axes[1, 0].set_xticklabels([bin_labels[i] for i in range(0, I, 4)], rotation=45)

# Bottom right: Regime used per bin
axes[1, 1].step(range(I), regimes_used, 'steelblue', linewidth=2, where='mid')
axes[1, 1].set_xlabel("Bin")
axes[1, 1].set_ylabel("Regime")
axes[1, 1].set_title("Regime Assignment Through the Day")
axes[1, 1].set_yticks(range(model_a.regime_classifier.n_regimes))
axes[1, 1].set_xticks(range(0, I, 4))
axes[1, 1].set_xticklabels([bin_labels[i] for i in range(0, I, 4)], rotation=45)

plt.tight_layout()
plt.show()

print(f"Day {walkthrough_day}: MAPE = {np.mean(errors):.1f}%")
print(f"Worst bin: {bin_labels[np.argmax(errors)]} ({errors[np.argmax(errors)]:.1f}%)")
print(f"Best bin:  {bin_labels[np.argmin(errors)]} ({errors[np.argmin(errors)]:.1f}%)")
````

## 6. Results and Validation

Now let's evaluate the model on 63 out-of-sample days. We compare Model A against the historical average baseline using MAPE (Mean Absolute Percentage Error):

$$\text{MAPE} = \frac{1}{N} \sum_{d,i} \frac{|\hat{V}_{d,i} - V_{d,i}|}{V_{d,i}}$$

A lower MAPE means better forecasts. The MAPE reduction tells us how much the model improves over the baseline:

$$\text{MAPE Reduction} = \frac{\text{MAPE}_{\text{baseline}} - \text{MAPE}_{\text{model}}}{\text{MAPE}_{\text{baseline}}} \times 100\%$$

````python
# Evaluate Model A on AAPL out-of-sample
eval_days = list(range(train_end, n_days))

model_mape = compute_evaluation_mape(model_a, vm, eval_days, params.min_volume_floor)
baseline_mape = compute_baseline_mape(vm, eval_days, params.N_hist, params.min_volume_floor)
reduction = (baseline_mape - model_mape) / baseline_mape * 100

print(f"=== {TICKER} Out-of-Sample Results (63 days) ===")
print(f"Baseline MAPE (historical average): {baseline_mape:.4f} ({baseline_mape*100:.1f}%)")
print(f"Model A MAPE:                       {model_mape:.4f} ({model_mape*100:.1f}%)")
print(f"MAPE Reduction:                     {reduction:.1f}%")

# Per-bin MAPE comparison
per_bin_model = np.zeros(I)
per_bin_baseline = np.zeros(I)
per_bin_count = np.zeros(I)

for d_idx in eval_days:
    h_start = max(0, d_idx - params.N_hist)
    hist_avg_d = np.mean(vm[h_start:d_idx], axis=0)
    
    for i in range(I):
        actual = vm[d_idx, i]
        if actual < params.min_volume_floor:
            continue
        observed = {j+1: vm[d_idx, j] for j in range(i)}
        predicted = forecast_raw_volume(model_a, vm, d_idx, i+1, observed)
        per_bin_model[i] += abs(predicted - actual) / actual
        per_bin_baseline[i] += abs(hist_avg_d[i] - actual) / actual
        per_bin_count[i] += 1

mask = per_bin_count > 0
per_bin_model[mask] /= per_bin_count[mask]
per_bin_baseline[mask] /= per_bin_count[mask]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: Per-bin MAPE comparison
axes[0].plot(range(I), per_bin_baseline * 100, 'b--o', markersize=4, label='Baseline', alpha=0.7)
axes[0].plot(range(I), per_bin_model * 100, 'r-s', markersize=4, label='Model A', alpha=0.7)
axes[0].set_xlabel("Bin")
axes[0].set_ylabel("MAPE (%)")
axes[0].set_title(f"{TICKER}: Per-Bin MAPE (Model vs Baseline)")
axes[0].legend()
axes[0].set_xticks(range(0, I, 4))
axes[0].set_xticklabels([bin_labels[i] for i in range(0, I, 4)], rotation=45)

# Right: Per-bin MAPE reduction
per_bin_reduction = np.zeros(I)
for i in range(I):
    if per_bin_baseline[i] > 0:
        per_bin_reduction[i] = (per_bin_baseline[i] - per_bin_model[i]) / per_bin_baseline[i] * 100

axes[1].bar(range(I), per_bin_reduction, color='seagreen', alpha=0.7)
axes[1].axhline(0, color='black', linestyle='-', linewidth=0.5)
axes[1].set_xlabel("Bin")
axes[1].set_ylabel("MAPE Reduction (%)")
axes[1].set_title(f"{TICKER}: Per-Bin MAPE Reduction (higher = better)")
axes[1].set_xticks(range(0, I, 4))
axes[1].set_xticklabels([bin_labels[i] for i in range(0, I, 4)], rotation=45)

plt.tight_layout()
plt.show()

print(f"\nPer-bin reduction ranges from {per_bin_reduction.min():.1f}% to {per_bin_reduction.max():.1f}%")
print("Improvement increases through the day as intraday ARMA gains more conditioning data.")
````

## 7. Limitations and Extensions

````python
# Show examples where the model struggles
# Find worst and best days in evaluation period
day_mapes = []
for d_idx in eval_days:
    day_ape = 0
    count = 0
    for i in range(I):
        actual = vm[d_idx, i]
        if actual < params.min_volume_floor:
            continue
        observed = {j+1: vm[d_idx, j] for j in range(i)}
        predicted = forecast_raw_volume(model_a, vm, d_idx, i+1, observed)
        day_ape += abs(predicted - actual) / actual
        count += 1
    day_mapes.append(day_ape / max(count, 1))

day_mapes = np.array(day_mapes)
worst_idx = np.argmax(day_mapes)
best_idx = np.argmin(day_mapes)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for ax_idx, (label, rel_idx) in enumerate([("Best Day", best_idx), ("Worst Day", worst_idx)]):
    d_idx = eval_days[rel_idx]
    actual_d = vm[d_idx]
    fc_d = []
    for i in range(I):
        observed = {j+1: vm[d_idx, j] for j in range(i)}
        fc_d.append(forecast_raw_volume(model_a, vm, d_idx, i+1, observed))
    
    axes[ax_idx].plot(range(I), actual_d, 'ko-', linewidth=2, markersize=4, label='Actual', alpha=0.7)
    axes[ax_idx].plot(range(I), fc_d, 'rs-', linewidth=2, markersize=4, label='Forecast', alpha=0.7)
    axes[ax_idx].set_xlabel("Bin")
    axes[ax_idx].set_ylabel("Volume")
    axes[ax_idx].set_title(f"{label} (MAPE = {day_mapes[rel_idx]*100:.1f}%)")
    axes[ax_idx].legend()
    axes[ax_idx].set_xticks(range(0, I, 4))
    axes[ax_idx].set_xticklabels([bin_labels[i] for i in range(0, I, 4)], rotation=45)

plt.tight_layout()
plt.show()

print(f"\nDay-level MAPE distribution across {len(eval_days)} evaluation days:")
print(f"  Mean:   {np.mean(day_mapes)*100:.1f}%")
print(f"  Median: {np.median(day_mapes)*100:.1f}%")
print(f"  Best:   {np.min(day_mapes)*100:.1f}%")
print(f"  Worst:  {np.max(day_mapes)*100:.1f}%")
print(f"  Std:    {np.std(day_mapes)*100:.1f}%")
````

### Known Limitations

1. **Model B's deviation constraint is too conservative.** The fixed 10% cap on the surprise regression adjustment limits the model's ability to adapt percentage forecasts. The paper uses proprietary adaptive bounds. Widening to 15-20% or using multiplicative adjustments could help.

2. **Computational cost.** Training takes ~90 seconds per stock, primarily due to the ARMA grid search (36 inter-day combinations x 25 intraday). Evaluation is expensive because each bin's forecast re-applies the full ARMA model history. A Kalman filter implementation would make online prediction O(1) per bin.

3. **Intraday ARMA day-boundary handling.** The concatenated-segment approach for fitting is an approximation. Ideally, each day would be treated as an independent segment with state resets at boundaries. The current approach may pick up spurious cross-day dependencies, though the short segment length (26 bins) limits this.

4. **Stationarity assumption.** The model assumes the volume process is approximately stationary within each fitting window. Stocks undergoing major structural changes (like INTC's 2024-2025 restructuring) violate this assumption and produce poor forecasts.

5. **No cross-sectional information.** Each stock is modeled independently. Market-wide volume shocks (e.g., FOMC announcements, options expiration) affect all stocks simultaneously, and a cross-sectional model could capture these systematic patterns.

### Potential Extensions

- **Adaptive deviation bounds** for Model B, learned from recent forecast accuracy
- **Kalman filter** state updates for O(1) online prediction
- **Cross-sectional features** (market-wide volume, VIX, sector volume) as additional components
- **Event-day detection** (earnings, FOMC, options expiry) with separate regime handling
- **Finer time resolution** (5-minute or 1-minute bins) for more precise VWAP execution
